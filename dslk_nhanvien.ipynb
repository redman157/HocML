```
import sys

class Office:
    def __init__(self, id, name):
        self.id = id
        self.name = name

class Employee:
    def __init__(self, id, name, salary, office):
        self.id = id
        self.name = name 
        self.salary = salary
        self.office = office

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self._head = None
        self._tail = None
        self._size = 0

    @property
    def size(self):
        return self._size

    def push(self, record):
        node = Node(record)
        node.next = None

        if self._tail is None:
            self._tail = node
        else:
            self._tail.next = node
            self._tail = node
        if self._head is None:
            self._head = node
        self._size += 1
 
    def show(self):
        curr = self.head
        while curr != None:
            print(curr.data)
            curr = curr.next
       
    def check(self, callback):
        curr = self._head

        while curr != None:
            if callback(curr.data) is True:
                return False

            curr = curr.next
        else:
            return True

    @property
    def head(self):
        return self._head

    @property
    def tail(self):
        return self._tail
    
    def get(self, index):
        curr = self._head

        for _ in range(0, index):
            if curr is None:
                return None
            else:
                curr = curr.next
        else:
            return curr.data

def convert(gmtime):
    hour, minute = gmtime.split(':')
    return int(hour) + float(minute)/60.0

def time_by_step(enter, exit, barrier):
    if enter > exit:
        raise AssertionError("enter({}) <= exit({}) but not".format(enter, exit))

    return exit - enter if exit - enter < barrier - enter else barrier - enter

def is_time(time):
    return 0 <= time and time <= 24

def is_id_existed(list, id):
    def callback(data):
        return data.id == id

    return not list.check(callback)

def menu(): 
    print("Menu:")

    while True:
        print("a/ Thêm nhân viên mới")
        print("b/ Thêm phòng ban mới")
        print("c/ Hiển thị danh sách nhân viên xuất sắc")
        print("d/ Kết thúc")
        print("")

        result = input("Nhập lệnh thực thi:")
        if result == "a":
            return 1
        elif result == "b":
            return 2
        elif result == "c":
            return 3
        elif result == "d":
            return 4
        else:
            print("Làm ơn kiểm tra lại mã lệnh và nhập lại theo hướng dẫn sau")

def main():
    llist_offices = LinkedList()
    llist_employees = LinkedList()

    while True:
        cmd = menu()

        if cmd == 4:
            print("Cám ơn đã sử dụng dịch vụ của chúng tôi, tạm biệt")
            sys.exit(0)
        elif cmd == 3:
            curr_office = llist_offices.head

            
            print("ID\tTên phòng ban\tMã nhân viên\tTên nhân viên\tLương")
            while curr_office != None:
                curr_employee = llist_employees.head
                rec_office = curr_office.data
                best_first_employee = None

                while curr_employee != None:
                    rec_employee = curr_employee.data
                    if rec_employee.office == rec_office.id:
                        if best_first_employee is None or best_first_employee.salary < rec_employee.salary:
                            best_first_employee = rec_employee
                    curr_employee = curr_employee.next
                else:
                    if not best_first_employee is None:
                        print("%s\t%s\t%s\t%s\t%d" % (rec_office.id, rec_office.name, best_first_employee.id, best_first_employee.name, best_first_employee.salary))
     
                curr_employee = llist_employees.head
                best_second_employee = None
                while curr_employee != None:
                    rec_employee = curr_employee.data
 
                    if rec_employee.office == rec_office.id:
                        if best_second_employee is None or best_second_employee.salary < rec_employee.salary:
                            if rec_employee.id != best_first_employee.id:
                                best_second_employee = rec_employee
                    curr_employee = curr_employee.next
                else: 
                    if not best_second_employee is None:
                        print("%s\t%s\t%s\t%s\t%d" % (rec_office.id, rec_office.name, best_second_employee.id, best_second_employee.name, best_second_employee.salary))
                curr_office = curr_office.next
        elif cmd == 2:
            retry = ""
            n = int(input("Nhập số lượng phòng ban: "))

            for i in range(n):
                print("Nhập thông tin phòng ban thứ %d:" % i)

                while True:
                    id = input("\tNhập %smã phòng ban: " % retry)

                    if is_id_existed(llist_offices, id) is False:
                        break
                    else:
                        retry = "lại "
                else:
                    retry = ""

                name = input("\tNhập tên phòng ban: ")
                llist_offices.push(Office(id, name))
        elif cmd == 1:
            retry = ""

            if llist_offices.size == 0:
                print("Nhập phòng ban trước khi nhập nhân viên")
                continue
            n = int(input("Nhập số lượng nhân viên: "))
            for i in range(n):
                print("Nhập thông tin nhân viên thứ %d:" % i)

                while True:
                    id = input("\tNhập %smã nhân viên: " % retry)

                    if is_id_existed(llist_employees, id) is False:
                        break
                    else:
                        retry = "lại "
                retry = ""

                name = input("\tNhập tên nhân viên: ")
                salary = int(input("\tNhập lương của nhân viên %s:" % name))

                while True:
                    office = input("\tNhập %smã phòng ban của nhân viên %s" % (retry, name))

                    if is_id_existed(llist_offices, office) is True:
                        break
                    else:
                        retry = "lại "

                llist_employees.push(Employee(id, name, salary, office))

if __name__ == '__main__':
    main()
```
